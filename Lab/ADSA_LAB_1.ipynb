```python
class Node:
    def __init__(self, name, age, patient_id):
        self.name = name
        self.age = age
        self.patient_id = patient_id
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None

    # Insert patient at end
    def insert_patient(self, name, age, patient_id):
        new_node = Node(name, age, patient_id)

        if self.head is None:
            self.head = new_node
            print("\n Patient inserted as first record.")
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_node
        print("\n Patient inserted successfully.")


    def delete_patient(self, patient_id):
        if self.head is None:
            print("\nâš ï¸ List is empty. Nothing to delete.")
            return


        if self.head.patient_id == patient_id:
            self.head = self.head.next
            print("\nðŸ—‘ï¸ Patient deleted successfully.")
            return

        current = self.head
        while current.next and current.next.patient_id != patient_id:
            current = current.next

        if current.next is None:
            print("\nâŒ Patient ID not found. Deletion failed.")
        else:
            current.next = current.next.next
            print("\nðŸ—‘ï¸ Patient deleted successfully.")


    def display_patients(self):
        if self.head is None:
            print("\nðŸ“­ No patients in the list.")
            return

        current = self.head
        print("\nðŸ“‹ Current Patient List:")
        print("--------------------------------------")
        while current:
            print(f"Name: {current.name}, Age: {current.age}, ID: {current.patient_id}")
            current = current.next
        print("--------------------------------------")



linked_list = SinglyLinkedList()

while True:
    print("\n===== Patient Management Using Singly Linked List =====")
    print("1. Insert Patient")
    print("2. Delete Patient (by ID)")
    print("3. Display Patient List")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        name = input("\nEnter Patient Name: ")
        age = int(input("Enter Patient Age: "))
        patient_id = int(input("Enter Patient ID: "))
        linked_list.insert_patient(name, age, patient_id)

    elif choice == '2':
        delete_id = int(input("\nEnter Patient ID to delete: "))
        linked_list.delete_patient(delete_id)

    elif choice == '3':
        linked_list.display_patients()

    elif choice == '4':
        print("\n Exiting Program...")
        break

    else:
        print("\n Invalid choice! Please enter a valid option.")
```
