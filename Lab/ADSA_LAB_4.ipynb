```python
# Towers of Hanoi using Recursive Approach
def towers_of_hanoi(n, source, auxiliary, destination):
    """
    Recursive function to solve Towers of Hanoi.
    n : Number of disks
    source : The source rod
    auxiliary : The helper rod
    destination : The destination rod
    """

    if n == 1:
        print(f"Move disk 1 from {source} â†’ {destination}")
        return

    towers_of_hanoi(n - 1, source, destination, auxiliary)

    print(f"Move disk {n} from {source} â†’ {destination}")

    towers_of_hanoi(n - 1, auxiliary, source, destination)

print("===== Towers of Hanoi using Recursion =====")
num_disks = int(input("Enter the number of disks: "))

print(f"\nðŸ§© Steps to solve Towers of Hanoi with {num_disks} disks:\n")
towers_of_hanoi(num_disks, 'A', 'B', 'C')
```
