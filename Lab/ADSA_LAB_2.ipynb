```python
# Circular Linked List for Patient Round-Robin Check-up System
class Node:
    def __init__(self, name, age, patient_id):
        self.name = name
        self.age = age
        self.patient_id = patient_id
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Insert patient at end
    def insert_patient(self, name, age, patient_id):
        new_node = Node(name, age, patient_id)

        if self.head is None:  # First patient
            self.head = new_node
            new_node.next = self.head
            print("\nâœ… Patient inserted as the first record.")
            return

        current = self.head
        while current.next != self.head:  # Traverse till last
            current = current.next

        current.next = new_node
        new_node.next = self.head
        print("\nâœ… Patient inserted successfully.")

    # Delete patient via patient ID
    def delete_patient(self, patient_id):
        if self.head is None:
            print("\nâš ï¸ List is empty. Nothing to delete.")
            return

        current = self.head
        prev = None

        # If first node matches
        if current.patient_id == patient_id:
            # If only one node in Circular Linked List
            if current.next == self.head:
                self.head = None
                print("\nğŸ—‘ï¸ Patient deleted successfully.")
                return

            # Move to last node to update pointer
            while current.next != self.head:
                current = current.next

            current.next = self.head.next
            self.head = self.head.next
            print("\nğŸ—‘ï¸ Patient deleted successfully.")
            return

        # For other nodes
        current = self.head
        while current.next != self.head:
            prev = current
            current = current.next
            if current.patient_id == patient_id:
                prev.next = current.next
                print("\nğŸ—‘ï¸ Patient deleted successfully.")
                return

        print("\nâŒ Patient ID not found. Deletion failed.")

    # Display circular patient list
    def display_patients(self):
        if self.head is None:
            print("\nğŸ“­ No patients in the list.")
            return

        current = self.head
        print("\nğŸ”„ Round-Robin Patient List:")
        print("----------------------------------------")

        while True:
            print(f"Name: {current.name}, Age: {current.age}, ID: {current.patient_id}")
            current = current.next
            if current == self.head:
                break

        print("-----------------------------------------")


linked_list = CircularLinkedList()

while True:
    print("\n===== Circular Linked List Patient System =====")
    print("1. Insert Patient")
    print("2. Delete Patient (by ID)")
    print("3. Display Patient List")
    print("4. Exit")

    choice = input("Enter your choice (1â€“4): ")

    if choice == '1':
        name = input("\nEnter Patient Name: ")
        age = int(input("Enter Patient Age: "))
        patient_id = int(input("Enter Patient ID: "))
        linked_list.insert_patient(name, age, patient_id)

    elif choice == '2':
        delete_id = int(input("\nEnter Patient ID to delete: "))
        linked_list.delete_patient(delete_id)

    elif choice == '3':
        linked_list.display_patients()

    elif choice == '4':
        print("\n Exiting Program...")
        break

    else:
        print("\n Invalid choice! Please enter a valid option.")
```
