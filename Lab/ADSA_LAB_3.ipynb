```python
class UndoRedoSystem:
    def __init__(self):
        self.undo_stack = []
        self.redo_stack = []


    def push_action(self, action):
        self.undo_stack.append(action)
        self.redo_stack.clear()
        print(f"\n‚úÖ Action performed: {action}")

    def undo(self):
        if not self.undo_stack:
            print("\n‚ö†Ô∏è No actions to undo.")
            return

        action = self.undo_stack.pop()
        self.redo_stack.append(action)
        print(f"\n‚Ü©Ô∏è Undo performed: {action}")

    def redo(self):
        if not self.redo_stack:
            print("\n‚ö†Ô∏è No actions to redo.")
            return

        action = self.redo_stack.pop()
        self.undo_stack.append(action)
        print(f"\nüîÅ Redo performed: {action}")

    def display(self):
        print("\nüìã ----- CURRENT STATE -----")
        print(f"Undo Stack: {self.undo_stack}")
        print(f"Redo Stack: {self.redo_stack}")
        print("-----------------------------")


system = UndoRedoSystem()

while True:
    print("\n===== Undo-Redo System Using Two Stacks =====")
    print("1. Perform Action (Push)")
    print("2. Undo Action")
    print("3. Redo Action")
    print("4. Display Stacks")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        action = input("\nEnter action to perform: ")
        system.push_action(action)

    elif choice == '2':
        system.undo()

    elif choice == '3':
        system.redo()

    elif choice == '4':
        system.display()

    elif choice == '5':
        print("\n Exiting Program...")
        break

    else:
        print("\n Invalid choice! Enter a valid option.")
```
